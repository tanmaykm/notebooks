```python
import numpy as np
from pandas import Series, DataFrame
import pandas as pd
from datetime import datetime
import matplotlib as plt
```

```python
# time series with date range of frequency 1 minute
rng = pd.date_range('2012-07-17', periods=100000, freq='t')
ts = Series(np.random.randn(100000), index=rng)
ts.head()
```

```python
ts[ts.index[:15]]
```

```python
# generate a daterange with frequency 30 minutes
rng30 = pd.date_range(ts.index[0], ts.index[-1], freq='30t')
# get data points every 30 minutes
ts[rng30]
```

```python
# resample at 5000 minutes with default mean aggregation
ts5000 = ts.resample('5000t')
ts5000
```

```python
# resample back to higher frequency
ts5000.resample('1000t')
```

```python
# forward fill the NAs
ts5000.resample('1000t').fillna(method='ffill', limit=2)
```

```python
# backward fill the NAs
ts5000.resample('1000t').fillna(method='bfill', limit=2)
```

```python
# forward with interpolation
hf = ts5000.resample('1000t')
hf
```

```python
hf.interpolate(method='values')
```

```python
# resample and count (closed can be left or right)
ts.resample('5000t', how='count', closed='right', label='right')
```

```python
# passing custom function is slower
```

```python
timeit ts.resample('30t', how='mean')
```

```python
timeit ts.resample('30t', how=lambda x: x.mean())
```

```python
# we can pass multiple functions
ts.resample('30t', how=['mean', 'median', 'max'])
```
